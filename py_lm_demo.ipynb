```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

Create a mock dataset

```python
# Generate random data
np.random.seed(0)
X = np.random.rand(100, 1)
y = 2 + 3 * X + np.random.rand(100, 1)
```

```python

# Visualize the data
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title('Random Data haha')
plt.show()
```

Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Create and train the linear regression model 

```python
# Create the model
linear_regression = LinearRegression()

# Train the model
linear_regression.fit(X_train, y_train)
```

Display Model's results

```python
print("Intercept (b0):", linear_regression.intercept_)
print("Coefficient (b1):", linear_regression.coef_)
```

Make Prediction and evaluate the model 

```python
# Make predictions on the test set
y_pred = linear_regression.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

Display the fitted regression line

```python
plt.scatter(X, y)
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
